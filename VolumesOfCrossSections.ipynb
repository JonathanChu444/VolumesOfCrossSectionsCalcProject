```python
import numpy as np
import matplotlib.pyplot as plt

"""
Preface:
    I coded this program to demonstrate volumes using circular
    and semicicular cross sections to better visualize how they
    look as I found it tough to understand what their shape was
    in a 3D space with X, Y, and Z axes because I have never graphed
    in 3D for math before calculus.
"""

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

#Below is the code for volumes of revolution in a known cross section between two functions.
theta = np.linspace(0, 2 * np.pi, 100)

#These five lines below will ask you in text boxes beneath all of the code for the top and
#bottom functions that create the base of the 3D Shape along with whether or not you want
#circular or semicircular cross sections. The fourth and fifth line asks for the bounds of the shape.
topFunction = input("What will be the top function?: ")
bottomFunction = input("What will be the bottom function?: ")
semiCircle = input("Type and enter Y if you want to have semicirclular cross sections or N if you want to have circular cross sections (case sensitive): ")
boundA = int(input("What will the first bound of the shape be?: "))
boundB = int(input("What will the second bound of the shape be?: "))

#The line below sets the bounds of the function
x = np.linspace(boundA, boundB, 100)

thetas, xs = np.meshgrid(theta, x)


if(semiCircle == "Y"):
    semiCircleToggle = True
elif(semiCircle == "N"):
    semiCircleToggle = False

def y(x):
    #The line below sets the radius between the two equations, the top one and the bottom one.
    radius = ( eval(topFunction) - eval(bottomFunction) ) / 2

    # This line below creates the Y coordinate of the circle.
    doneY = radius * np.cos(thetas) + radius

    return doneY

def z(x):
    #The line below sets the radius between the two equations, the top one and the bottom one.
    radius = ( eval(topFunction) - eval(bottomFunction) )/ 2

    #This line below creates the Z coordinate of the circle.
    doneZ = radius * np.sin(thetas)

    #The 2 lines below sets the limit of whether or not you want semicircle cross sections or not
    if(semiCircleToggle):
        doneZ[doneZ < 0] = np.nan

    return doneZ

ax.plot_surface(xs, y(xs), z(xs))

plt.show()
```
